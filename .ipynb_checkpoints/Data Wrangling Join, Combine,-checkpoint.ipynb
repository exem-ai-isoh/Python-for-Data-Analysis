## Combining and Merging Datasets, key 기준으로 병합

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
df1
df2
```

```python
# key 컬럼을 기준으로 merge
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
# 중복되는 key가 없는 경우 직접 지정 가능하다, 기본적으로 inner join
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

## Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])

pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

## Concatenating Along an Axis, 축 따라 이어붙이기

```python
arr = np.arange(12).reshape((3, 4))
arr
# Numpy는 concatenate함수로 이어붙이기 기능 제공
np.concatenate([arr, arr], axis=1)
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

```python
# Pandas는 concat 함수로 이어붙이기 기능 제공
pd.concat([s1, s2, s3])
```

```python
pd.concat([s1, s2, s3], axis=1)
```

## Combining Data with Overlap

병합이나 이어붙이기로는 데이터를 합치는 것이 불가능한 상황이 있는데 두 데이터의 색인이 일부 혹은 전체가 겹치는 경우 그러하다. 

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b[-1] = np.nan
a
```

```python
b
```

```python
# Numpy의 where 함수는 if - else 구문 표현한다.
np.where(pd.isnull(a), b, a)
```

```python
# 위와 동일한 기능 + 정렬
b[:-2].combine_first(a[2:])
```

## Pivoting “Long” to “Wide” Format

```python
data = pd.read_csv('examples/macrodata.csv')
data.head()
```

```python
# PeriodIndex 시간 간격을 나타내기 위한 자료형, 연도와 분기 컬럼을 합친다.
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter,
                         name='date')
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D', 'end')
ldata = data.stack().reset_index().rename(columns={0: 'value'})

# RDBMS 관점에서 date, item은 primary key에 해당한다. 아래의 형식은 질의에 있어 강점이 있다. 하지만 데이터가 너무 길어 작업이 용이하지 못할 수 있다.
ldata
```

```python
# pivot 메서드에서 처음 두 인자는 로우와 컬럼 색인으로 사용될 컬럼 이름이고 마지막 두 인자는 DF에 채워 넣을 값을 담고 있는 컬럼 이름
pivoted = ldata.pivot('date', 'item', 'value')
pivoted
```

```python
ldata['value2'] = np.random.randn(len(ldata))
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item')
pivoted[:5]
```

```python
pivoted['value'][:5]
```

## Pivoting “Wide” to “Long” Format

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])
melted
```

```python
reshaped = melted.pivot('key', 'variable', 'value')
reshaped
```

```python
reshaped.reset_index()
```

```python
# 사용할 컬럼들의 집합을 지정
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
```
